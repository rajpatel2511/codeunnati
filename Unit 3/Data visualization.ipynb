```python
# x axis
x =[5,2,9,4,7,3,8,10,1]
#y axis
y = [10,5,8,4,2,1,6,3,9]

plt.plot(x,y,"g")
plt.xlabel("x")
plt.ylabel("y")

#here we sat the for ticks,rotation and color value

plt.tick_params(axis="x",labelsize=18,labelrotation=-60,labelcolor="r")
plt.tick_params(axis="y",labelsize=16,labelrotation=20,labelcolor="blue")

plt.show()
```

# Adding annotatae and legend

```python
import numpy as np

import matplotlib.pyplot as plt

#you can also write
#from matplotlib import pylot as plt
```

```python
x = np.arange(0,10)
y=x^2
z=x^3

plt.plot(x,y)
plt.plot(x,z)
plt.plot(y,z)
plt.title("Graph")
plt.xlabel("x")
plt.ylabel("y")




#Annotation
plt.annotate(xy=[2,1], text="First Entry",rotation=20)
plt.annotate(xy=[4,6], text="Second Entry",rotation=20)
plt.annotate(xy=[8,10], text="Third Entry",rotation=20)

#adding legend
plt.legend(["blue","orange","green"])

plt.show()
```

# Pie Chart

pie chart is a circular plot that can display only one series of data.

he area of chart is the total % of the given data

```python
#import libraries
import numpy as np

import matplotlib.pyplot as plt
```

```python
#creating dataset

Branch = ["computer","Electrical","mechnical","civil","automobile","IT"]

Data = [50,34,40,29,12,23]

explode=(0.10,0,0,0,0.1,0)

#creat chart

fig = plt.figure(figsize=(10,7))
plt.pie(Data, labels=Branch, explode=explode, autopct="%1.1f%%",shadow=True)
plt.show()
```

# 3D Ploting

```python
#import mplot3d toolkit ,numpy,& matplotlib
import numpy as np

import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d #3d toolkit lib

```

```python
fig = plt.figure()
#syntax for 3d projection

ax = plt.axes(projection = "3d")
#defining all 3 axes

z=np.linspace(0,1,100)

x = z*np.sin(25*z)
y = z*np.cos(25*z)

#3d ploting

ax.plot3D(x,y,z,"green")

ax.set_title("3D line plotine")
plt.show()
```
